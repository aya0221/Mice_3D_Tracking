## Extracting Frames from .bag File
### Extracts color and depth frames from the original .bag file and saves them as PNG images

(This process takes long time)

```python
# Path to your .bag file
bag_file = '/data/recording_data/20231103_162601.bag'

# Path for color and depth frames extracted from .bag file
color_frames_dir = '/data/color_depth/color_frames'
depth_frames_dir = '/data/color_depth/depth_frames'
os.makedirs(color_frames_dir, exist_ok=True)
os.makedirs(depth_frames_dir, exist_ok=True)

# Configure pipeline to read from the .bag file
pipeline = rs.pipeline()
config = rs.config()
rs.config.enable_device_from_file(config, bag_file)

# Start the pipeline to get the stream profile
try:
    profile = pipeline.start(config)
except Exception as e:
    print("Error starting pipeline:", e)
    exit(1)

# Get the stream profile and configure the streams dynamically
stream_profile = profile.get_streams()
for sp in stream_profile:
    # DEPTH EXTRACTION
    if sp.stream_type() == rs.stream.depth:
        config.enable_stream(rs.stream.depth, sp.as_video_stream_profile().width(), 
                             sp.as_video_stream_profile().height(), rs.format.z16, sp.fps())
    # COLOR EXTRACTION
    elif sp.stream_type() == rs.stream.color:
        config.enable_stream(rs.stream.color, sp.as_video_stream_profile().width(), 
                             sp.as_video_stream_profile().height(), rs.format.rgb8, sp.fps())

# Stop the pipeline once in order to restart the pipeline with the correct configuration
pipeline.stop()
try:
    pipeline.start(config)
except Exception as e:
    print("Error restarting pipeline with new configuration:", e)
    pipeline.stop()
    exit(1)

try:
    frame_number = 0
    while True:
        frames = pipeline.wait_for_frames()
        color_frame = frames.get_color_frame()
        depth_frame = frames.get_depth_frame()

        # Ignore other data section
        if not color_frame or not depth_frame:
            continue

        color_image = np.asanyarray(color_frame.get_data())
        depth_image = np.asanyarray(depth_frame.get_data())

        color_img_path = os.path.join(color_frames_dir, f'color_frame_{frame_number}.png')
        depth_img_path = os.path.join(depth_frames_dir, f'depth_frame_{frame_number}.png')
        cv2.imwrite(color_img_path, color_image)
        cv2.imwrite(depth_img_path, depth_image)

        frame_number += 1

except Exception as e:
    print("Error processing frames:", e)
finally:
    pipeline.stop()
```

# Pick 400 random images

### Adjust Image Brightness and Contrast

In my code, the values for alpha and beta are set to 1.0 and 0 respectively to start with the original image's contrast and brightness. 

- Alpha (Contrast Control): between 1.0 (no change) and 3.0 (increased contrast).

- Beta (Brightness Control): between 0 (no change in the brightness of the image) and 100 

```python
# Adjust brightness and contrast of a sample image
alpha = 1.0  # Contrast control (1.0-3.0)
beta = 0     # Brightness control (0-100)
adjusted_image = cv2.convertScaleAbs(new_image, alpha=alpha, beta=beta)
plt.imshow(adjusted_image[:, :, [2, 1, 0]])  # Convert BGR to RGB
plt.show()


# ===AUTHOR's CODE===
# new_image = sampled_c_frames[500,...][:,:,[2,1,0]]
# # * Enter the alpha value [1.0-3.0]: 2.2
# alpha = 1#.5
# # * Enter the beta value [0-100]: 50
# beta = 0
# pic = cv2.convertScaleAbs(new_image, -1, alpha, beta)
# plt.imshow(pic)
```

## K-means Clustering
## k-means to group similar images in clusters

- class: 12

```python
# Apply K-means clustering on the grayscale images
from sklearn.cluster import MiniBatchKMeans
from sklearn.utils.validation import check_is_fitted
Machine = MiniBatchKMeans(n_clusters=12, max_iter=1000, batch_size=100, verbose=1)
# flatten everything, except the first dimension
g_frames_flat = np.reshape(sampled_g_frames, (sampled_g_frames.shape[0], -1))
Machine.fit(g_frames_flat)
cluster_labels = Machine.predict(g_frames_flat)


# ===AUTHOR's CODE===
# do k-means on the greyscale images!
# from sklearn.cluster import MiniBatchKMeans
# from sklearn.utils.validation import check_is_fitted

# Machine = MiniBatchKMeans(n_clusters=12, init='k-means++', max_iter=1000,
#                  batch_size=100, verbose=1, compute_labels=True,
#                  random_state=None, tol=0.0, max_no_improvement=10,
#                  init_size=None, n_init=10, reassignment_ratio=0.01)

# # flatten everything, except the first dimension
# g_frames_flat = np.reshape(sampled_g_frames,(sampled_g_frames.shape[0],-1))
# # downsample here?
# Machine.fit(g_frames_flat)
# cluster_labels = Machine.predict(g_frames_flat)
# sampled_g_frames = sampled_g_frames.astype('uint8')
# plt.imshow(sampled_g_frames[0,:,:]
```
